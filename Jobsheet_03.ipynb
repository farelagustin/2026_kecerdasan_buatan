1. Central Tendency

a. Mendapatkan Nilai Mean

```python
import numpy as np

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.mean(score)
print(x)
```

b. Mendapatkan Nilai Median

```python
import numpy as np

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.median(score)
print(x)
```

c. Mendapatkan Nilai Modus / Mode

```python
from scipy import stats

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = stats.mode(score, keepdims=True)
print(x)
```

2. Spread

a. Mendapatkan Nilai Range

```python
import numpy as np

data = [2, 3, 7, 8, 9, 12, 15, 20, 50, 42]

minimum = np.min(data)
maximum = np.max(data)
x = maximum - minimum
print(f"Nilai Minimum:{minimum}")
print(f"Nilai Maximum:{maximum}")
print(f"NIlai Range  :{x}")
```

b. Mendapatkan Nilai Percentile

```python
import numpy as np

data = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.percentile(data, 90)
print(x)
```

c. Mendapatkan Nilai Quartile dan Interquartile Range

```python
import numpy as np

data = [3, 4, 4, 5, 5, 6, 6, 7, 8, 8, 9, 10]

Q1 = np.median(data[:int(len(data)/2)])
Q3 = np.median(data[int(len(data)/2):])
IQR = Q3 - Q1

print(f"Nilai Quartile 1: {Q1}")
print(f"Nilai Quartile 3: {Q3}")
print(f"Nilai Interquartile Range: {IQR}")
```

d. Mendapatkan Nilai Variance dan Standart Deviation

```python
import numpy as np

data= [86, 87, 88, 86, 87, 85, 86]

mean = np.mean(data)
var = np.var(data)
std = np.std(data)

print(f"Nilai Mean: {mean}")
print(f"Nilai Variance: {var}")
print(f"Nilai Standart Deviation: {std}")
```

Tugas Praktikum 03

```python
import numpy as np

data = [50, 54, 62, 50, 52, 59, 61, 63, 65, 10, 53, 63, 65, 50, 59, 62, 50, 51, 57, 60, 63, 65, 65, 53, 99]

data_sorted = sorted(data)

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"Data Nilai Hasil Ujian(Terurut dari yang terkecil sampai terbesar): {data_sorted}")
print("-" * 30)
print(f"Nilai Quartile 1: {Q1}")
print(f"Nilai Quartile 3: {Q3}")
print(f"Nilai Interquartile Range: {IQR}")
print("-" * 30)
print(f"Lower Bound    : {lower_bound}")
print(f"Upper Bound    : {upper_bound}")
print("-" * 30)
print(f"Outlier ditemukan: {outliers}")
```

Tugas Praktikum 04

```python
import numpy as np
import pandas as pd

data = {
    'Berat Badan (kg)': ['41-45', '46-50', '51-55', '56-60', '61-65', '66-70'],
    'Frekuensi (f)': [2, 3, 6, 4, 5, 2]
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Mencari Titik Tengah (xi) untuk setiap interval
def get_midpoint(interval):
    start, end = map(int, interval.split('-'))
    return (start + end) / 2

df['xi'] = df['Berat Badan (kg)'].apply(get_midpoint)

# Menghitung f * xi untuk mencari rata-rata
df['f_xi'] = df['Frekuensi (f)'] * df['xi']

total_f = df['Frekuensi (f)'].sum()
total_f_xi = df['f_xi'].sum()
rata_rata = total_f_xi / total_f

# Menghitung Varians dan Standar Deviasi
# Rumus: s = sqrt( sum(f * (xi - x_bar)^2) / n )
df['xi_minus_mean_sq'] = (df['xi'] - rata_rata)**2
df['f_xi_minus_mean_sq'] = df['Frekuensi (f)'] * df['xi_minus_mean_sq']

varians = df['f_xi_minus_mean_sq'].sum() / total_f
standar_deviasi = np.sqrt(varians)

print("Tabel Perhitungan:")
print(df[['Berat Badan (kg)', 'Frekuensi (f)', 'xi', 'f_xi']])
print("-" * 45)
print(f"Total Frekuensi (n)    : {total_f}")
print(f"Rata-rata (Mean)       : {rata_rata:.2f} kg")
print(f"Standar Deviasi (s)    : {standar_deviasi:.2f}")
```
