3. Menggunakan Fungsi MATH

```python
# Menghitung Sinus
print(math.sin(math.pi/2))
# Menghitung kosinus
print(math.cos(0))
# Menghitung tangen
print(math.tan(math.radians(45)))
# Mengubah sudut dari radian ke derajat.
print(math.degrees(math.pi))

# Menghitung logaritma natural (basis $e$) secara default, atau basis tertentu.
print(math.log(10))     # Logaritma natural dari 10
print(math.log(8, 2))   # Logaritma basis 2 dari 8
# Menghitung logaritma basis 10.
print(math.log10(100))

# Menghitung jarak Euclidean antara dua titik (misal titik A ke titik B).
titik1 = (0, 0)
titik2 = (3, 4)
print(math.dist(titik1, titik2))

# Menghitung sisa bagi (modulo) untuk bilangan desimal (float).
print(math.fmod(20, 3))
print(math.fmod(5.5, 2))

# Menghitung nilai faktorial (n!).
print(math.factorial(5)) # 5*4*3*2*1

# Menghitung permutasi (jumlah cara memilih $k$ objek dari $n$ tanpa pengulangan dan memperhatikan urutan).
print(math.perm(5, 2))

# Menghitung fungsi Gamma (sering disebut faktorial yang diperluas untuk bilangan desimal, di mana Gamma(n) = (n-1)!).
print(math.gamma(5)) # Sama dengan factorial(4)

# Mengecek apakah sebuah nilai adalah NaN (Not a Number/Bukan Angka).
angka_aneh = float('nan')
print(math.isnan(angka_aneh))
```

4. Operasi matriks dan vektor dengan python

a. Membuat matriks di python

```python
import numpy as np
A = np.array([[80, 30, 25],
              [50, 20, 10]])
B = np.array([[30, 60, 15],
              [10, 30, 30]])

hasil = A + B
print(hasil)
```

b. Melakukan operasi dot product

```python
import numpy as np
A = np.array([10, 20, 15])
B = np.array([16, 67, 35])

hasil = np.dot(A, B)
print(hasil)
```

c. Melakukan operasi cross product

```python
import numpy as np
A = np.array([10, 20, 15])
B = np.array([16, 67, 35])

Hasil = np.cross(A, B)
print(Hasil)
```

Tugas Praktikum 01

```python
import numpy as np
A = np.array([[3, 1],
             [2, 0]])
B = np.array([[2, 1],
              [1, -1]])
C = np.array([[-1, 1, 2],
              [2, 0, -1]])
# A x B
a = np.cross(A, B)
# B . C
b = np.dot(B, C)

print(f"a. A x B = {a}")
print(f"b. B . C = \n {b} \n")
```

Tugas Praktikum 02

```python
import math

# 6!
soal_a = math.factorial(6)
# 8! / 5!
soal_b = math.factorial(8) / math.factorial(5)

print(f"a. 6! = {soal_a}")
print(f"b. 8! / 5! = {soal_b}")
```

Tugas Praktikum 03

```python
import numpy as np
A = np.array([5, 3, 7])
B = np.array([12, -3, 1])

# Menentukan Panjang Masing-masing Vektor (Norm)
panjang_A = np.linalg.norm(A)
panjang_B = np.linalg.norm(B)

# Perkalian Titik (A.B)
dot_product = np.dot(A, B)

# Perkalian Silang (AxB)
cross_product = np.cross(A, B)

print(f"Panjang Vektor A: {panjang_A:.2f}")
print(f"Panjang Vektor B: {panjang_B:.2f}")
print(f"Hasil Perkalian Titik (A.B): {dot_product}")
print(f"Hasil Perkalian Silang (AxB): {cross_product}")
```
